# Check trajectory coordinates in FIDASIM

```python
%matplotlib ipympl
import matplotlib.pyplot as pl
import numpy as np
import picklr
```

```python
pl.close('all')
```

```python
# function to read the FIDASIM dictionary
import pickle as PCL

#read a pickled dictionary

def read_dict(filename):
    f = None
    with open(filename, 'rb') as f:
        d = PCL.load(f)
        f.close()
    return d
```

```python
# load the FD test file
d = read_dict('Test_detector_array1.pcl')
# get the sightlines array
ds = d['sightline']

# print the shape
print(f'Sightline array shape : {ds.shape}')
print(f'nenergy: {d["nenergy"]}, coords: 6, nsteps: {d["nsteps"]}, nrays: {d["nrays"]}, ndet: {d["nchan"]}') 
```

```python
# get the 1st energy trajectory
t = ds[0,:,:,0,0]
```

```python
# get coordinates
# cyl. coords
r_r = t[3,:]
is_nz = r_r > 0.

npts = is_nz.sum ()

r = t[3,:npts];phi=t[4,:npts];z_r = t[5,:npts]

vr = t[0,:npts];vphi = t[1,:npts];vz = t[2,:npts]


# convert to cart. 
x = r*np.cos(phi); y = r*np.sin(phi)
vx = vr*np.cos(phi) - vphi*np.sin(phi)
vy = vr*np.sin(phi) + vphi*np.cos(phi)
vxy = np.sqrt(vx**2 + vy**2)

```

```python
# check that velocities are constant
V_mag = np.sqrt(vx**2 + vy**2 + vz**2)
```

## load temp output FIDASIM trajectory output file

```python

df = np.loadtxt('fort.26')
xx = df[:npts,0]
yy = df[:npts,1]
zz = df[:npts,2]

vxx = df[:npts,3]
vyy = df[:npts,4]
vzz = df[:npts,5]

vxxyy = np.sqrt(vxx**2 + vyy**2)
```

```python
# get the 1st energy trajectory B-fields
tb = dsb[0,0,0,:,:]
```

```python
br = tb[:,0][:npts] 
bphi = tb[:,1][:npts]
bz = tb[:,2][:npts]
b_tot = tb[:,3][:npts]
psi_rel = tb[:,4][:npts]

```

```python
# convert to cartesian

# r-component
rv_xy = np.array([x,y, np.zeros_like(x)])
rv_xy_mag = np.linalg.norm(rv_xy, axis = 0)
u_r = rv_xy/rv_xy_mag

# phi component
u_phi = np.array([-u_r[1,:], u_r[0,:], np.zeros_like(x)])

u_z = np.array([np.zeros_like(x),np.zeros_like(x),np.ones_like(x)])

Bf_c = br*u_r + bphi*u_phi + bz * u_z
Bx = Bf_c[0]
By = Bf_c[1]
Bz = Bf_c[2]
```
